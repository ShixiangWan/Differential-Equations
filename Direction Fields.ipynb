# Direction Fields In Python

We are going to use the libraries numpy and matplotlib. We'll render plots inline. 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 16, 12 # makes the plots bigger
```

We'll start by looking at the direction field for the differential equation $\dfrac{dy}{dx} = x\cdot y$. First, we need to specify the region on which we wish to sketch the direction field.

```python
x = np.linspace(-2, 2, 32)
y = np.linspace(-1.5, 1.5, 24)
xm, ym = np.meshgrid(x, y) # this forms the grid of points based on the ranges specified above
```

Next we define the lineal elements. Since the equation is in standard form $\dfrac{dy}{dx} = f(x, y)$, the slope at any point $(x_0, y_0)$ is just $f(x_0, y_0)$. Slope is defined as vertical change per unit change in $x$. So for each lineal element, change in $x$, or $dx$, is $1$, and change in $y$, or $dy$, is $f(x, y) = x\cdot y$.

```python
dx = 1
dy = xm*ym
```

```python
plt.quiver(xm, ym, dx, dy, pivot='middle', headwidth=2, headlength=3, color='blue')
plt.plot(x , -0.5*np.exp(x**2/2), color='red', linewidth=2.25) # plots solution curve through (0, -0.5) 
plt.plot(x, 0.3*np.exp(x**2/2), color='green', linewidth=2.25) # plots solution curve through (0, 0.3)
plt.ylim((-1.5, 1.5))
```

Next let's look at $\dfrac{dy}{dx} = \sin(y)$. All the steps below are analagous to those above: first we specify the region, then we specify the change in $y$ for each lineal element (note that change in $x$ is still $1$, so we don't need to redefine it), then we plot.

```python
x_2 = np.linspace(-5, 5, 25)
y_2 = np.linspace(-5, 5, 25)
xm_2, ym_2 = np.meshgrid(x_2, y_2)
dy_2 = np.sin(ym_2)
plt.quiver(xm_2, ym_2, dx, dy_2, pivot='middle', headwidth=2, headlength=3, color='red')
```

Finally, let's look at $\dfrac{dy}{dx} = \sin(x)\cdot \cos(y)$. I'm going to plot on the same region, so the only thing to redefine is the slope. 

```python
dy_3 = np.sin(x_2)*np.cos(y_2)
plt.quiver(x_2, y_2, dx, dy_3, pivot='middle', headwidth=2, headlength=3, color='magenta')
```
